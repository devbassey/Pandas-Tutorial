# PANDAS

Pandas lets you load, explore, and analyze data using Python code. It makes the job faster than clicking on Excel sheets..

Pandas gives you two super powers.
- A series --> one column of data (basically a list with labels)
- A DataFrame --> a full table (rows + columns, just like a full spreadsheet of data)

To use Python in a Notebook, you have to import pandas as pd

```python
# import pandas library into the notebook
import pandas as pd # the "as pd"part is just a nickname. So instead of typing pandas everytime, you type pd
```

## Python Series

A Series is like a single column in a spreadsheet. It is simply a list of values with an index (row labels).

```python
# A simple series of student scores
scores = pd.Series([85, 92, 78, 95, 60], index=['alice', 'bob', 'carol', 'david', 'eve'])

scores
```

## Python DataFrame

A DataFrame is a full spreadsheet with multiple columns and multiple rows.

Almost everything you will do in Data Analysis starts with a DataFrame.

```python
# Example: A Class Grade Book

data = {
    'Name' : ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Math' : [85, 92, 78, 95, 60],
    'Science' : [88, 79, 95, 85, 72],
    'English' : [90, 85, 80, 88, 65]
}

df = pd.DataFrame(data) # we convert the dictionary to dataFrame

df
```
