```python
# Creating a Series Using Dictionary

fruit_protein = {
    'Avovado': 2.0,
    'Guava': 2.6,
    'Blackberries': 2.0,
    'Oranges': 0.9,
    'Banana': 1.1,
    'Apples': 0.3,
    'Kiwi': 1.1,
    'Udara': 1.7,
    'Mango': 0.8,
    'Cherries': 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name = 'Protein')
```

```python
s2
```

```python
# Conditional Selection (s2 greater than 1)

s2>1
```

```python
s2[s2>1]
```

```python
# Logical Operators (and, or, not)

[(s2>0.5) & (s2<2)]
```

```python
s2[(s2>0.5) & (s2<2)] # finding only the value that satisfy this condition
```

```python
s2[(s2>0.5) | (s2<2)] # Logical OR (|) operation
```

```python
# Logical NOT Operation

s2[~(s2>1)] # finds all the values that are not greater than the value 1
```

```python
s2[~(s2<1)] # finds all the values that are not less than the value 1
```

```python
# MOdifying the series

s2['Mango'] = 2.8 # Modifies the value of mango from 0.8 to 2.8
```

```python
s2
```

# DATAFRAME DATA STRUCTURE IN PANDAS

```python
# DataFrame

data = {
    'Name': ['Utibe', 'Bassey', 'Inyang', 'Henry', 'Happiness', 'Hopex'],
    'Age':[30, 25, 20, np.nan, 29, 25],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'HR'],
    'Salary': [50000, 60000, 70000, 62000, np.nan, 50000]
}
```

```python
data
```

```python
# Converting the Python Dictionary (data) to a DataFrame
df = pd.DataFrame(data)
df
```

```python
df.head(2) # --> Displays the first two rows
```

```python
df.tail(2) # Displays the last two rows
```

```python
# Fetching the first and Second row in the dataset using the iloc and loc indexing

df.iloc[0:3]
```

```python
df.loc[0:3, ['Name','Age', 'Department']]
```

```python
df['Age'] # queries a single column
```

```python
df[['Age', 'Department']]
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Broadcasting in pandas

df['Salary'] = df['Salary'] + 5000

df['Salary']
```

```python
df
```

```python
# Renaming Columns in a DataFrame

df.rename(columns={'Department': 'Dept'}, inplace= True)
```

```python
df
```

```python
df['Dept'].unique()
```

```python
# HOw many employees in each Department
df['Dept'].value_counts() # Counts the values in a dataset
```

```python
df['Promoted Salary'] = df['Salary'] * 10
df
```
