# PANDAS

Pandas is a Python Library built ontop of Numpy. Widely used in DA,DS and ML.

The name Pandas come from Panel Data and has nothing to do with pandas -  the animal

In Pandas, you typically work with Series and DataFrame.
- A series is a 1-dimensional label
- A dataFrame is a 2 dimensional labelled grid or table



```python
import pandas as pd
```

```python
student = {
    'name' : ['utibe', 'john'],
    'age' : [12, 23],
    'department' : ['it', 'customer care']
}
```
